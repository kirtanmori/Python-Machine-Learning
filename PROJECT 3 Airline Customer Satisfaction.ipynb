```python
data.drop("id",axis=True,inplace=True)
```

```python
x=data.drop("satisfaction",axis=1)
```

```python
y=data["satisfaction"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=50)
```

```python
model1=StandardScaler()
```

```python
x_train=model1.fit_transform(x_train)
```

```python
x_test=model1.transform(x_test)
```

```python
pd.DataFrame(x_train)
```

```python
data
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(criterion="gini",splitter="best",max_depth=12,max_features="sqrt")
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import precision_score,recall_score,accuracy_score,classification_report,confusion_matrix,f1_score
```

```python
confusion_matrix(y_test,model.predict(x_test))
```

```python
print(classification_report(y_test,model.predict(x_test)))
```

```python
print(precision_score(y_test,model.predict(x_test)))
print(recall_score(y_test,model.predict(x_test)))
print(accuracy_score(y_test,model.predict(x_test)))
print(f1_score(y_test,model.predict(x_test)))
```
