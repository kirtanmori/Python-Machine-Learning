```python
from sklearn.cluster import KMeans
```

```python
model.fit(data)
```

```python
data["cluster"]=model.predict(data)
```

```python
df1=data[data["cluster"]==0]
df2=data[data["cluster"]==1]
df3=data[data["cluster"]==2]
```
