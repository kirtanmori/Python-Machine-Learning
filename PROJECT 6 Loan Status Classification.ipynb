```python
data["Default Status"].replace(['Default', 'Not Default'],[0,1],inplace=True)
```

```python
data
```

```python
x=data.drop("Default Status",axis=1)
```

```python
y=data["Default Status"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=10)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,classification_report
```

```python
print(accuracy_score(y_test,model.predict(x_test)))
print(precision_score(y_test,model.predict(x_test)))
print(recall_score(y_test,model.predict(x_test)))
print(confusion_matrix(y_test,model.predict(x_test)))   
print(classification_report(y_test,model.predict(x_test)))
```
