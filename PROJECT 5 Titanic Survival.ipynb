```python
data.drop(data[data["Embarked"].isnull()].index,inplace=True)
```

```python
data["Age"]=data["Age"].fillna(30)
```

```python
data.drop("Ticket",axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc=LabelEncoder()
```

```python
data[["Sex","Embarked"]]=data[["Sex","Embarked"]].apply(enc.fit_transform)
```

```python
data
```

```python
data.drop(["Name","Cabin"],axis=1,inplace=True)
```

```python
x=data.drop("Survived",axis=1)
```

```python
y=data["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score,classification_report
```

```python
confusion_matrix(y_test,model.predict(x_test))
```

```python
print(classification_report(y_test,model.predict(x_test)))
```

```python
print(accuracy_score(y_test,model.predict(x_test)))
print(precision_score(y_test,model.predict(x_test)))
print(recall_score(y_test,model.predict(x_test)))
print(f1_score(y_test,model.predict(x_test)))
```
